```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from image_analysis.generate_image import *
from image_analysis.image_synthetic import *
from image_analysis.put_text import *
```

```python
# 鉛筆
height = 405
width = 320
top_size = 151
pencil = cv2.imread("../input_image/top_1_altered.png")
pencil_resized = cv2.resize(pencil, (top_size, top_size))
background = generate_white(height, width)
x_location = 80
y_location = 50
pencil_syn = synthetic(background, pencil_resized, [y_location, x_location])
pencil_syn = pencil_syn.astype(int)
plt.imshow(pencil_syn)
```

```python
pencil = draw_vertical_line(pencil_syn, [0, 0, 0], 0)
pencil = draw_horizontal_line(pencil, [0, 0, 0], 404)
pencil = draw_horizontal_line(pencil, [0, 0, 0], 0)
pencil = draw_vertical_line(pencil, [0, 0, 0], 319)
```

```python
image = cv2_put_text(pencil, "登録する", (30, 270), '/System/Library/Fonts/ヒラギノ角ゴシック W4.ttc' , 70, (0, 0, 0))
plt.imshow(image)
```

```python
cv2.imwrite("../output_image/pencil.png", image)
```
